```python
def equation(x0, t):
    theta, x = x0
    f = [x, -(g/l)*np.sin(theta)]
    return f

def plot_results(tempo, theta1, theta2):
    plt.plot(tempo, theta1[:,0])
    plt.plot(tempo, theta2)
    plt.title("Movimento Pendular")
    plt.xlabel('t(s)')
    plt.ylabel('angulo')
    plt.grid(True)
    #plt.legend
```

```python
g = 9.81
l=1.0
tempo = np.arange(0, 10.0, 0.025)
```

```python
initial_angle = 45.0
theta0 = np.radians(initial_angle)
x0 = np.radians(0.0)
```

```python
theta1 = odeint(equation, [theta0, x0], tempo)
```

```python
w = np.sqrt(g/l)
theta2 = [theta0* np.cos(w*t) for t in tempo]
```

```python
plot_results(tempo, theta1, theta2)
```
